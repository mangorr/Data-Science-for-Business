```python
import sys
import os
import pandas as pd
import numpy as np

file_dir="/Users/yachiru/Documents/nyu/class/DSforBusinessTech/assignment/assignment1/stock_data/"
all_file_list=os.listdir(file_dir)

dic = {}

for single_file in all_file_list:
    if ".csv" in single_file:
        single_data_frame=pd.read_csv(os.path.join(file_dir,single_file))
        
        dic[single_file[:-4]] = single_data_frame
```

```python
# Assignment7
for key, value in dic.items():
    value[key] = (value['Adj Close'].diff() / value['Adj Close'].shift(1)).fillna(0)
    display(value)
```

```python
# Assignment8

mdf_intersect = dic["AAL"].iloc[:, [0,-1]]

for key, value in dic.items():
    print(key)
    if key == "AAL":
        continue
    else:
            mdf_intersect = mdf_intersect.merge(dic[key][['Date',key]], 
                         how = 'inner', on = 'Date')
display(mdf_intersect)
```

```python
# Assignment9
mdf_intersect["sum_returns"] = mdf_intersect.sum(axis = 1)
display(mdf_intersect)

spy_data_frame=pd.read_csv("/Users/yachiru/Documents/nyu/class/DSforBusinessTech/assignment/assignment1/stock_data/spy/SPY.csv")
spy_data_frame["SPY"] = (spy_data_frame['Close'].diff() / spy_data_frame['Close'].shift(1)).fillna(0)

display(spy_data_frame)

mdf_intersect = mdf_intersect.merge(spy_data_frame.iloc[:, [0, -1]], how = 'inner', on = 'Date')
display(mdf_intersect)

corr = mdf_intersect["sum_returns"].corr(mdf_intersect["SPY"])
print(corr)
```

```python
# Assignment10
for i in range(1, len(mdf_intersect.columns)):
    column_name = mdf_intersect.columns.values[i]
    corr_by_column = mdf_intersect.iloc[:, i].corr(mdf_intersect["SPY"])
    startDate = "2019-06-20"
    endDate = "2019-08-31"
    
    print(f'The correlation between SPY and {column_name} is {100 * corr_by_column:.2f}% (start date : {startDate}. end date : {endDate})')
```

```python
# Assignment12
mdf_union = dic["AAL"].iloc[:, [0,-1]]
display(dic["AAL"])

for key, value in dic.items():
    if key == "AAL":
        continue
    else:
        mdf_union = mdf_union.merge(dic[key][['Date',key]], 
                         how = 'outer', on = 'Date')

# mdf_union = mdf_union.merge(spy_data_frame.iloc[:, [0, -1]], how = 'outer', on = 'Date')
        
display(mdf_union)

result = 0
        
for i in range(0, len(mdf_union)):
    count = 0
    for j in range(1, len(mdf_union.columns)):
        if float(mdf_union.iloc[i, j]) > 0.0:
            count = count + 1
    
    if count >= 90:
        result = result + 1

print(result)
```
